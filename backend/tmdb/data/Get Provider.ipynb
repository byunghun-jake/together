```python
import json
import pandas as pd
```

```python
with open('./movies_kr.json', 'r') as f:
    data = json.loads(f.read())
```

```python
df = pd.json_normalize(data)
df = df.fillna(0)
```

```python
df.head()
```

```python
for idx, row in df.iterrows():
    if row['fields.provider.buy'] != 0:
        print(row['pk'])
        print(row['fields.provider.buy'][0]['provider_name'])    
    if row['fields.provider.rent'] != 0:
        print(row['fields.provider.rent'][0]['provider_name'])
    if row['fields.provider.flatrate'] != 0:
        print(row['fields.provider.flatrate'][0]['provider_name'])
```
